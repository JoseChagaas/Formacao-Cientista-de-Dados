```python
import pandas as pd
```

```python
#Importar planilha
dataset = pd.read_csv("dados.csv", sep=';')
dataset.head(10)
```

```python
#Verificar tipos de dados e valores nulos
dataset.info()
```

```python
#Estatisticas descritivas das colunas numericas
dataset.describe()
```

```python
#Analisand duplicatas na coluna 'MUNICIPIO'
municipio_duplicado = dataset.groupby("MUNICIPIO").size()
municipio_duplicado[municipio_duplicado > 1]
```

```python
#Reomovendo duplicatas na coluna 'MUNICIPIO'
dataset = dataset.drop_duplicates(subset="MUNICIPIO", keep="first")
```

```python
#Verificando se ficou alguma duplicata
municipio_duplicado = dataset.groupby("MUNICIPIO").size()
municipio_duplicado[municipio_duplicado > 1]
```

```python
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
#Criando histograma da distribuição de PIB
dataset['PIB'].hist(bins=20)
plt.title('Distribuição de PIB')
plt.xlabel('PIB')
plt.ylabel('Frequência')
plt.show()

#Criando histograma da distribuição de VALOREMPENHO
dataset['VALOREMPENHO'].hist(bins=20)
plt.title('Distribuição de Valor Empenho')
plt.xlabel('Valor Empenho')
plt.ylabel('Frequência')
plt.show()
```

```python
#Correlação entre as colunas numéricas da tabela
correlacao = dataset.corr(numeric_only=True)
correlacao
```

```python
#Criando gráfico de calor para visualização de correlação entre variaveis
plt.figure(figsize=(8,6))
sbn.heatmap(correlacao, annot=True, cmap="Blues", fmt=".2f")
plt.title("Mapa de Calor - Correlação entre Variáveis")
plt.show()
```

```python
#Gráfico de dispersão para visualizar correlação entre as variveis 
plt.figure(figsize=(8,6))
sbn.scatterplot(data=dataset, x='PIB', y='VALOREMPENHO')
plt.title('Dispersão entre PIB e Valor Empenhado')
plt.xlabel('PIB')
plt.ylabel('Valor Empenhado')
plt.grid(True)
plt.show()
```
