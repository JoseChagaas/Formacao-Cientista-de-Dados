```python
import pandas as pd
```

```python
#Importar planilha
dataset = pd.read_csv("dados.csv", sep=';')
dataset.head(10)
```

```python
#Verificar tipos de dados e valores nulos
dataset.info()
```

```python
#Estatisticas descritivas das colunas numericas
dataset.describe()
```

```python
#Analisand duplicatas na coluna 'MUNICIPIO'
municipio_duplicado = dataset.groupby("MUNICIPIO").size()
municipio_duplicado[municipio_duplicado > 1]
```

```python
#Reomovendo duplicatas na coluna 'MUNICIPIO'
dataset = dataset.drop_duplicates(subset="MUNICIPIO", keep="first")
```

```python
#Verificando se ficou alguma duplicata
municipio_duplicado = dataset.groupby("MUNICIPIO").size()
municipio_duplicado[municipio_duplicado > 1]
```

```python
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
#Criando histograma da distribuio de PIB
dataset['PIB'].hist(bins=20)
plt.title('Distribuio de PIB')
plt.xlabel('PIB')
plt.ylabel('Frequncia')
plt.show()

#Criando histograma da distribuio de VALOREMPENHO
dataset['VALOREMPENHO'].hist(bins=20)
plt.title('Distribuio de Valor Empenho')
plt.xlabel('Valor Empenho')
plt.ylabel('Frequncia')
plt.show()
```

```python
#Correlao entre as colunas numricas da tabela
correlacao = dataset.corr(numeric_only=True)
correlacao
```

```python
#Criando grfico de calor para visualizao de correlao entre variaveis
plt.figure(figsize=(8,6))
sbn.heatmap(correlacao, annot=True, cmap="Blues", fmt=".2f")
plt.title("Mapa de Calor - Correlao entre Variveis")
plt.show()
```

```python
#Grfico de disperso para visualizar correlao entre as variveis 
plt.figure(figsize=(8,6))
sbn.scatterplot(data=dataset, x='PIB', y='VALOREMPENHO')
plt.title('Disperso entre PIB e Valor Empenhado')
plt.xlabel('PIB')
plt.ylabel('Valor Empenhado')
plt.grid(True)
plt.show()
```
